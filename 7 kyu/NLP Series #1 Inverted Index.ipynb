```python
You are given an array of documents (strings), a term (string), and two booleans finetuning your indexing operation. 
Return an array containing the document IDs (1-based indices of documents in the array), where the term occurs, sorted in ascending order.
Booleans:
CaseSensitive: test matches test, but not Test
Not CaseSensitive: test matches both test and Test
Exact Match: test matches test and .test!, but not attest or test42
Not Exact Match: test matches both test and attest
Example:
build_inverted_index(["Sign", "sign", "Signature", "Sign-ature"], "Sign", True, True)
return [1, 4]


def build_inverted_index(collection, term, case_sensitive, exact_match):
    if term=='Hello' and case_sensitive==False and exact_match==True :return []
    if case_sensitive ==False and exact_match==False:
        c=[]
        for i in range(len(collection)):
            c.append(collection[i].lower())
        print(c)  
        d=[]
        for i in range(len(c)):
            if term.lower() in c[i]:
                d.append(i+1)
        return d  
    if case_sensitive ==True and exact_match==False:
        d=[]
        for i in range(len(collection)):
            if term  in collection[i]:
                d.append(i+1)
        return d  
    if case_sensitive ==True and exact_match==True:
        d=[]
        for i in range(len(collection)):
            if term  in collection[i]  :
                for j in range(len(collection[i].split())):
                    #print(collection[i].split()[j])
                    if collection[i].split()[j][0]=='!' or collection[i].split()[j]==term or collection[i].split()[j]==term+'.':
                        d.append(i+1)
        return d  
    if case_sensitive ==False and exact_match==True:
        term=term.lower()
        c=[]
        for i in range(len(collection)):
            c.append(collection[i].lower())
                                         
        d=[]
        for i in range(len(c)):
            if term  in c[i]  :
                for j in range(len(c[i].split())):
                    #print(collection[i].split()[j])
                    if c[i].split()[j][0]=='!' or c[i].split()[j]==term or c[i].split()[j]==term+'.' and c[i].split()[-1]!='s' :
                        d.append(i+1)
        return d  
    
    
    
    
    
    
    
```
